# Inference and Validation

네트워크를 학습시켰으면, 이제 예측을 해볼 수 있습니다. 이를 **insference**라고 하죠. 그런데 신경막은 트레이닝 데이터에 대해 너무 예측을 잘해서 그걸로 일반화로 하는 경향이 있습니다. **overfitting**이라고 하죠. 이는 inference 퍼포먼스를 저해합니다. overfitting을 태스팅하기 위해 우리는 training set이 아닌 다른 데이터에 대한 performance를 측정하는데 이를 **validation 셋**이라합니다. dropout 등 정규화를 통해 overfitting을 피합니다. 이 노트북에서, 당신에게 pytorch로 이걸 어떻게 하는지를 보여줄겁니다. 

datasets.DATA(train=False)로 둚으로써 test셋을 가져올 수 있습니다. 이는 나중에 더 배웁니다. 

```python
testset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=False, transform=transform)
```

테스트셋은 트레이닝셋의 10-20퍼센트에 해당하는 양의 이미지 데이터입니다. 

````python
import torch
from torchvision import datasets, transforms

# Define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.5, ), (0.5, ))])
# Download and load the training data
trainset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

# Download and load the test data
testset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=False, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=True)
````

Here I'll create a model like normal, using the same one from my solution for part 4.

````python
from torch import nn, optim
import torch.nn.functional as F

class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 10)
        
    def forward(self, x):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.log_softmax(self.fc4(x), dim=1)
        
        return x
````

validation의 목적은 데이터에 대한 모델의 퍼포먼스를 측정하는 것입니다. 트레이닝 셋에 포함되지 않은 데이터로요. 퍼포먼스는 정의하는 개발자에 달려있습니다. 전형적으로 이것은 accuracy입니다. 다른 방법으로는 [precision and recall](https://en.wikipedia.org/wiki/Precision_and_recall#Definition_(classification_context))과 top-5 error rate 등이 있습니다. 여기서는 accuracy에만 초점을 맞춥니다. forward pass부터 하나의 배치를 test set에서 할겁니다. 

````python
model = Classifier()

images, labels = next(iter(testloader))

# Get the class probabilities
ps = torch.exp(model(images))

# Make sure the shape is appropriate, we should get 10 class probabilities for 64 examples
print(ps.shape)
````

`ps.topk` 방법을 통해 가장 확률이 높은 class를 얻게됩니다. 이는 $k$개의 가장 높은 value를 리턴합니다. 우리는 가장 가능성 높은 class만을 원하기 때문에 `ps.topk(1)`를 사용합니다. 이는 top-$k$개의 값과 top-$k$ 인덱스를 가진 튜플을 리턴합니다. 만약 가장 높은 값이 5th 원소라면, index 4를 리턴합니다. 



````python
top_p, top_class = ps.topk(1, dim=1)
# Look at the most likely classes for the first 10 examples
print(top_class[:10,:])
````

이제 우리는 예측된 클래스가 label과 일치하는지를 체크합니다. 이것은 `top_class`와 `labels`를 equating함으로써 간단히 해결할 수 있습니다. 그러나 그 모양을 잘 살펴봐야합니다. `top_class`는 2d tensor로, 64 * 1 크기지만 `labels`는 1d이고 64크기 입니다. 둘을 비교하려면 같은 모양으로 맞춰야함. 


If we do

```python
equals = top_class == labels
```

`equals`는 `(64,64)` 모양일겁니다. `top-class`의 각 row의 요소를 `labels`의 각 요소와 비교하는 것은 64개 차원의 T/F boolean 값을 각 row에 리턴합니다. 

````python
equals = top_class == labels.view(*top_class.shape)
````

올바른 예측의 비율을 계산할 필요가 있습니다. `equals`는 binary value를 가지고 있습니다. 0또는 1이죠. 이는 우리가 모든 값을 더한다음 그 개수로 나누었을때, 올바른 prediction의 비율을 알아낼수있단겁니다. 이는 mean과 같아서, `torch.mean`을 불러서 accuracy를 계산하면됩니다. `torch.mean(equals)`를 하면 error를 알 수 있습니다. 

```
RuntimeError: mean is not implemented for type torch.ByteTensor
```

이는 equals가 torch.bytetensor의 타입인데 torch.mean이 그 타입의 tensor로 일을 못하기 때문입니다. 그래서 equals를 float으로 바꿔줘야합니다. torch.mean은 scalar입니다. accuray.item()을 통해 float을 리턴합니다. 

````python
accuracy = torch.mean(equals.type(torch.FloatTensor))
print(f'Accuracy: {accuracy.item()*100}%')
````

네트워크는 아직 학습안되었으므로 랜덤한 추측을 하고 있고 10퍼 남짓한 정확도를 보입니다. 이제 네트워크를 학습하고 validation pass를 포함시켜서 퍼포먼스를 측정할 수 있습니다. parameter를 업데이팅하고 있으므로, `torch.no_grad()`를 통해 gradient를 turn off하여 speed up 가능합니다. 


```python
# turn off gradients
with torch.no_grad():
    # validation pass here
    for images, labels in testloader:
        ...
```

>**Exercise:** Implement the validation loop below. You can largely copy and paste the code from above, but I suggest typing it in because writing it out yourself is essential for building the skill. In general you'll always learn more by typing it rather than copy-pasting.

````python
model = Classifier()
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr=0.003)

epochs = 30
steps = 0

train_losses, test_losses = [], []
for e in range(epochs):
    running_loss = 0
    for images, labels in trainloader:
        
        optimizer.zero_grad()
        
        log_ps = model(images)
        loss = criterion(log_ps, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    else:
        test_loss = 0
        accuracy = 0
        
        # Turn off gradients for validation, saves memory and computations
        with torch.no_grad():
            
            for images, labels in testloader:
                log_ps = model(images)
                test_loss += criterion(log_ps, labels)
                
                ps = torch.exp(log_ps)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == labels.view(*top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
                
        train_losses.append(running_loss/len(trainloader))
        test_losses.append(test_loss/len(testloader))

        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(running_loss/len(trainloader)),
              "Test Loss: {:.3f}.. ".format(test_loss/len(testloader)),
              "Test Accuracy: {:.3f}".format(accuracy/len(testloader)))
````

````python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
````

````python
plt.plot(train_losses, label='Training loss')
plt.plot(test_losses, label='Validation loss')
plt.legend(frameon=False)
````

## Overfitting


<img src='assets/overfitting.png' width=450px>

네트워크는 트레이닝 셋을 점점 더 잘 학습하여 loss가 줄어듭니다. 그러나 트레이닝 셋 바깥의 데이터를 일반화하여 validation loss가 올라가는 문제를 보입니다. 딥러닝 모델의 목표는 새로운 데이터에 대한 예측을 잘하는 것이기 때문에 가능하면 validation loss가 낮아지게 하고 싶습니다. 한 옵션은 모델의 버전 중 loss가 가장 낮은 것을 이용하는 겁니다. *early-stopping*이라고 합니다. 실제에서는 , 모델을 자주 자주 저장해서 가장 낮은 loss의 모델을 선택합니다. 

overfitting을 줄이는 다른 방법은 *dropout*입니다. 랜덤하게 input unit을 drop하는 겁니다. 이는 weight사이의 정보를 공유하게 하여 새 데이터에 대한 일반화 능력을 향상합니다. Adding dropout in PyTorch is straightforward using the [`nn.Dropout`](https://pytorch.org/docs/stable/nn.html#torch.nn.Dropout) module.



```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 10)
        
        # Dropout module with 0.2 drop probability
        self.dropout = nn.Dropout(p=0.2)
        
    def forward(self, x):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)
        
        # Now with dropout
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))
        x = self.dropout(F.relu(self.fc3(x)))
        
        # output so no dropout here
        x = F.log_softmax(self.fc4(x), dim=1)
        
        return x
```

트레이닝에서는 드랍아웃을 써도, inference할 때는 전체 네트워크를 쓰고 싶다. 그래서 우리는 validatin이나 testing을 할때는 dropout을 끈다. 이를 위해 `model.eval()`을 쓴다. 이는 dropout probability가 0인 evaluation 모드로 바꾼다. 그리고 `model.train()`을 통해 트레이닝 모드로 되돌린다. 보통 validation loop의 패턴은 이렇게 생겼다. gradient를 끄고, evaluation mode로 바꾸고, loss를 계산하고, 다시 train 모드로 되돌린다. 


```python
# turn off gradients
with torch.no_grad():
    
    # set model to evaluation mode
    model.eval()
    
    # validation pass here
    for images, labels in testloader:
        ...

# set model back to train mode
model.train()
```

> **Exercise:** Add dropout to your model and train it on Fashion-MNIST again. See if you can get a lower validation loss.

````python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 10)

        # Dropout module with 0.2 drop probability
        self.dropout = nn.Dropout(p=0.2)

    def forward(self, x):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)

        # Now with dropout
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))
        x = self.dropout(F.relu(self.fc3(x)))

        # output so no dropout here
        x = F.log_softmax(self.fc4(x), dim=1)

        return x
````

````python
model = Classifier()
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr=0.003)

epochs = 30
steps = 0

train_losses, test_losses = [], []
for e in range(epochs):
    running_loss = 0
    for images, labels in trainloader:
        
        optimizer.zero_grad()
        
        log_ps = model(images)
        loss = criterion(log_ps, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    else:
        test_loss = 0
        accuracy = 0
        
        # Turn off gradients for validation, saves memory and computations
        with torch.no_grad():
            model.eval()
            for images, labels in testloader:
                log_ps = model(images)
                test_loss += criterion(log_ps, labels)
                
                ps = torch.exp(log_ps)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == labels.view(*top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
        
        model.train()
        
        train_losses.append(running_loss/len(trainloader))
        test_losses.append(test_loss/len(testloader))

        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(running_loss/len(trainloader)),
              "Test Loss: {:.3f}.. ".format(test_loss/len(testloader)),
              "Test Accuracy: {:.3f}".format(accuracy/len(testloader)))
````

````python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
````

````python
plt.plot(train_losses, label='Training loss')
plt.plot(test_losses, label='Validation loss')
plt.legend(frameon=False)
````
