```python
import torch
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
from keras.preprocessing.sequence import pad_sequences
import torch.nn as nn

from tqdm import tqdm, trange
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from transformers import BertTokenizer, BertConfig, BertAdam, BertForSequenceClassification
from transformers import DistilBertForSequenceClassification, DistilBertTokenizer
from transformers import XLNetForSequenceClassification, XLNetTokenizer
from transformers import RobertaForSequenceClassification, RobertaTokenizer

import random
random.seed(321)
np.random.seed(321)
torch.manual_seed(321)
torch.cuda.manual_seed(321)

% matplotlib inline
```

## Device 설정 (GPU)

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
n_gpu = torch.cuda.device_count()
torch.cuda.get_device_name(0)
```

## Prepare the Data

```python
def load_imdb_data(tokenizer, MAX_LEN=128, n=10000, roberta=False):
    df = pd.read_csv('imdb_master.csv', encoding='latin-1')
    df.drop(['Unnamed: 0'], axis = 1, inplace = True)
    df.drop(['file'], axis = 1, inplace = True)
    df_n=df.copy()
    df_n=df_n[df_n['label'].isin(['neg','pos'])]
    df_n['label']=[0 if x=='neg' else 1 for x in df_n['label']]
    train_df = df_n[df_n['type'] == 'train'][:int(n*0.5)].append(df_n[df_n['type']=='train'][-int(n*0.5):])
    train_texts = train_df['review'].tolist()
    train_labels = list(train_df['label'])
    if roberta==True:
        train_tokens = list(map(lambda t: ['<s>'] + tokenizer.tokenize(t)[:126] + ['</s>'], train_texts))
    else:
        train_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:126] + ['[SEP]'], train_texts))
    
    # Use the BERT tokenizer to convert the tokens to their index numbers in the BERT vocabulary
    train_ids = [tokenizer.convert_tokens_to_ids(x) for x in train_tokens]
    # Pad our input tokens
    train_ids = pad_sequences(train_ids, maxlen=MAX_LEN, dtype="long", truncating="post", padding="post")
    # Create attention masks
    train_masks = []
    # Create a mask of 1s for each token followed by 0s for padding
    for seq in train_ids:
      seq_mask = [float(i>0) for i in seq]
      train_masks.append(seq_mask)

    
    test_df = df_n[df_n['type'] =='test'][:500].append(df_n[df_n['type'] =='test'][-500:])
    test_texts = test_df['review'].tolist()
    test_labels = list(test_df['label'])
    if roberta==True:
        test_tokens = list(map(lambda t: ['<s>'] + tokenizer.tokenize(t)[:126] + ['</s>'], test_texts))
    else:
        test_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:126] + ['[SEP]'], test_texts))

    test_ids=[tokenizer.convert_tokens_to_ids(x) for x in test_tokens]
    test_ids=pad_sequences(test_ids, maxlen=MAX_LEN, dtype="long", truncating="post", padding="post")
    test_masks=[]
    for seq in test_ids:
      seq_mask = [float(i>0) for i in seq]
      test_masks.append(seq_mask)
    
    # Convert all of our data into torch tensors, the required datatype for our model
    train_ids = torch.tensor(train_ids)
    test_ids = torch.tensor(test_ids)
    train_labels = torch.tensor(train_labels)
    test_labels = torch.tensor(test_labels)
    train_masks = torch.tensor(train_masks)
    test_masks = torch.tensor(test_masks)
    
    return train_ids,train_masks,train_labels,test_ids,test_masks,test_labels


def make_dataloader(data_ids,data_masks,data_labels,batch_size):
    data=TensorDataset(data_ids,data_masks,data_labels)
    sampler=RandomSampler(data)
    dataloader=DataLoader(data,sampler=sampler,batch_size=batch_size)
    return dataloader

# Function to calculate the accuracy of our predictions vs labels
def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten() #예측한 수치 (소수점임)
    labels_flat = labels.flatten() #정답    
    return np.sum(pred_flat == labels_flat) / len(labels_flat)
```

## 1. Bert Model

```python
### Bert Model 

# Number of training epochs (authors recommend between 2 and 4)
batch_size = 30
epochs = 5

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
train_ids,train_masks,train_labels,test_ids,test_masks,test_labels=load_imdb_data(tokenizer)
train_dataloader=make_dataloader(train_ids,train_masks,train_labels,batch_size)
test_dataloader=make_dataloader(test_ids,test_masks,test_labels,batch_size)

# Load BertForSequenceClassification, the pretrained BERT model with a single linear classification layer on top.       
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
model = nn.DataParallel(model)
model.cuda()
param_optimizer = list(model.named_parameters())
no_decay = ['bias', 'gamma', 'beta']
optimizer_grouped_parameters = [
    {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.01},
    {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.0}
]
# This variable contains all of the hyperparemeter information our training loop needs
optimizer = BertAdam(optimizer_grouped_parameters,lr=2e-5, warmup=.1)
```

## Bert Train

```python
##Bert Model:Train

# Store our loss and accuracy for plotting
train_loss_set = []
loss_lst=[]
batch_cnt=0

# trange is a tqdm wrapper around the normal python range
for _ in trange(epochs, desc="Epoch"):
    print(time.strftime("%x %X", time.localtime()))
    batch_cnt+=1

    # Training

    # Set our model to training mode (as opposed to evaluation mode)
    model.train()

    # Tracking variables
    tr_loss = 0
    nb_tr_examples, nb_tr_steps = 0, 0

    # Train the data for one epoch
    for step, batch in enumerate(train_dataloader):
        
        # Add batch to GPU
        batch = tuple(t.to(device) for t in batch)
        
        # Unpack the inputs from our dataloader
        b_input_ids, b_input_mask, b_labels = batch
        
        # Clear out the gradients (by default they accumulate)
        optimizer.zero_grad()
        
        # Forward pass
        loss = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask, labels=b_labels)
        loss=torch.mean(loss)
        train_loss_set.append(loss.item())    
        
        # Backward pass
        loss.backward()
        
        # Update parameters and take a step using the computed gradient
        optimizer.step()


        # Update tracking variables
        tr_loss += loss.item()
        nb_tr_examples += b_input_ids.size(0)
        nb_tr_steps += 1

    print(time.strftime("%x %X", time.localtime()))
    loss_lst.append(tr_loss/nb_tr_steps)
    print("Train loss: {}".format(tr_loss/nb_tr_steps))
    
```

## Bert Loss

```python
## Bert Model

plt.figure(figsize=(15,8))
plt.title("Training loss")
plt.xlabel("Batch")
plt.ylabel("Loss")
plt.plot(train_loss_set)
plt.show()
```

## Bert Test

```python
#Bert Model

# Test set Validation

# Put model in evaluation mode to evaluate loss on the validation set
model.eval()

# Tracking variables 
eval_loss, eval_accuracy = 0, 0
nb_eval_steps, nb_eval_examples = 0, 0

# Evaluate data for one epoch
for batch in test_dataloader:
    # Add batch to GPU
    batch = tuple(t.to(device) for t in batch)
    # Unpack the inputs from our dataloader
    b_input_ids, b_input_mask, b_labels = batch

    # Telling the model not to compute or store gradients, saving memory and speeding up validation
    with torch.no_grad():
    # Forward pass, calculate logit predictions
        logits = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask)

    # Move logits and labels to CPU
    logits = logits.detach().cpu().numpy()
    label_ids = b_labels.to('cpu').numpy()

    tmp_eval_accuracy = flat_accuracy(logits, label_ids)
    eval_accuracy += tmp_eval_accuracy
    nb_eval_steps += 1


print("Validation Accuracy: {}".format(eval_accuracy/nb_eval_steps))
```

## 2. XLNet

```python
# Number of training epochs (authors recommend between 2 and 4)
batch_size = 20
epochs = 5

tokenizer = XLNetTokenizer.from_pretrained('xlnet-base-cased')

train_ids,train_masks,train_labels,test_ids,test_masks,test_labels=load_imdb_data(tokenizer, n=10000, roberta=True)
train_dataloader=make_dataloader(train_ids,train_masks,train_labels,batch_size)
test_dataloader=make_dataloader(test_ids,test_masks,test_labels,batch_size)

model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased")
model = nn.DataParallel(model)
model.cuda()


#====================================================
param_optimizer = list(model.named_parameters())
no_decay = ['bias', 'gamma', 'beta']
optimizer_grouped_parameters = [
    {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.01},
    {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.0}
]

optimizer = BertAdam(optimizer_grouped_parameters,lr=2e-5, warmup=.1)
```

### XLNet train

```python
# Store our loss and accuracy for plotting
train_loss_set = []

loss_lst=[]
batch_cnt=0

# trange is a tqdm wrapper around the normal python range
for _ in trange(5, desc="Epoch"):
    print(time.strftime("%x %X", time.localtime()))    

    # Training

    # Set our model to training mode (as opposed to evaluation mode)
    model.train()

    # Tracking variables
    tr_loss = 0
    nb_tr_examples, nb_tr_steps = 0, 0

    # Train the data for one epoch
    for step, batch in enumerate(train_dataloader):
        # Add batch to GPU
        batch = tuple(t.to(device) for t in batch)
        # Unpack the inputs from our dataloader
        b_input_ids, b_input_mask, b_labels = batch
        # Clear out the gradients (by default they accumulate)
        optimizer.zero_grad()
        # Forward pass
#         outputs = model(b_input_ids, attention_mask=b_input_mask, labels=b_labels)
        outputs = model(b_input_ids, labels=b_labels)
        loss = outputs[0]
#         logits = outputs[1]
        loss=torch.mean(loss)
        train_loss_set.append(loss.item())    
        # Backward pass
        loss.backward()
        # Update parameters and take a step using the computed gradient
        optimizer.step()


        # Update tracking variables
        tr_loss += loss.item()
        nb_tr_examples += b_input_ids.size(0)
        nb_tr_steps += 1
    batch_cnt+=1
        
    print(time.strftime("%x %X", time.localtime()))
    print("Train loss: {}".format(tr_loss/nb_tr_steps))
```
